<h1>Regressão Logística</h1>

<pre>
    A regressão logística é um tipo de análise de regressão, utilizada para predizer o resultado de uma variável          categórica (uma variável que pode adotar um número limitado de categoria) em função de outras variáveis indepedentes
    
    É útil para modelar a probabilidade de que um evento pode ocorrer em função de outros elementos.
    
    É um método de classificação 
    
    A função de regressão Logística recorre a qualquer valor e devolverá uma valor entre 0 e 1.
    
    Se o resultado é >= 0.5 a saída será 1
    Se o resultado é < 0.5, a saída é 0
</pre>

A predição serce para saber a problabilidade de acerto da predição:

**prediçao = (Positivos corretos + Negativos corretos)/Total**

A taxa de erro serve para saber a probabilidade de erro da predição

**Taxa de error = (Falsos positivos + Falsos Negativos)/total**

<h3><font color='red' size=7>Limpeza dos Dados</font></h3>

```python
sns.heatmap(df.isnull())
```

```python
#1ª estratégia modificar os valor nulos das idades pela média
sns.boxplot(x='Pclass', y='Age', data=df)
```

```python
#função para a idade média media

def idade_media(colunas):
    idade = colunas[0]
    classe = colunas[1] 
    if pd.isnull(idade):
        if classe == 1:
            return 38
        elif classe == 2:
            return 28
        else:
            return 25
    else:
        return idade
        
```

```python
df['Age']= df[['Age', 'Pclass']].apply(idade_media, axis=1)
```

```python
sns.heatmap(df.isnull())
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
sns.heatmap(df.isnull())
```

```python
df.head()
```

```python
df.drop(['Name', 'Ticket', 'PassengerId'], axis=1, inplace=True)
```

```python
df.head()
```

```python
def transfomando_sexo(coluna):
    if coluna == 'male':
        return 1
    else:
        return 0
```

```python
df['Sex'] = df['Sex'].apply(transfomando_sexo)
```

```python
#sexo = pd.get_dummies(df['Sex'], drop_first=True)
```

```python
#df = pd.concat([df, sexo], axis=1 )
```

```python
df.head()
```

```python
def transformar_embarque(coluna):
    if coluna == 'S':
        return 1
    else:
        return 0
```

```python
df['Embarked'] = df['Embarked'].apply(transformar_embarque)
```

```python
df.head()
```

<h1><b>Treinamento do Modelo</b></h1>

```python
df.columns
```

```python
#Objetivo
Y = df['Survived']
#Características
X = df.drop('Survived', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=45) 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelo = LogisticRegression()
```

```python
#Treinamento do modelo
modelo.fit(X_train, Y_train)
```

```python
predicao = modelo.predict(X_test)
```

```python
predicao
```

**Métricas**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, predicao))
```

**Matriz de Confusão**

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(Y_test, predicao))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Exemplos**

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'red')
plt.show()
```

```python
plt.plot(x,y, 'green')
plt.title('Exemplo')
plt.legend('Dados')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g')

plt.subplot(1,2,2)
plt.plot(y,x, 'red')

plt.show()
```

**Programação Orientada a Objetos**

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y, 'silver')
```

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,1,1])
grafico.plot(x,y, 'silver')
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y, 'blue')


grafico2 = figura.add_axes([0.2,0.6,0.3,0.3])
grafico2.plot(y,x, 'red')
```

**MultiGráficos**

```python
figura, grafico = plt.subplots(nrows=2, ncols=2)
```

```python
figura, grafico = plt.subplots(nrows=2, ncols=5)
```

```python
figura, grafico = plt.subplots(nrows=4, ncols=3)
```

```python
figura, grafico = plt.subplots(nrows=2, ncols=2)

grafico[0][0].plot(x,y, 'blue')
grafico[0][1].plot(y,x, 'red')
grafico[1][0].plot(x,y, 'green')
grafico[1][1].plot(y,x, 'yellow')

grafico[0][0].set_title('Azul')
grafico[0][1].set_title('Vermelho')
grafico[1][0].set_title('Verde')
grafico[1][1].set_title('Amarelo')
```

**Tamanho do Gráfico**

```python
figura = plt.figure(figsize=(8,4))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])

grafico.plot(x,y, 'yellow')
```

```python
figura = plt.figure(figsize=(20,10))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])

grafico.plot(x,y, 'yellow')
```

**Criando dois Gráficos Juntos**

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])

grafico1.plot(x, x**2, label='Valor ao Quadrado' )
grafico1.plot(x, x**3, label='Valor ao Cubo' )
grafico1.legend(loc=0)
```

**Tipos de linha dos Gráficos**

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x, y, linewidth=10, alpha=0.5)
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x, y, linestyle="--")
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x, y, linestyle=":")
```

```python
figura = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x, y, linestyle="-.")
```
